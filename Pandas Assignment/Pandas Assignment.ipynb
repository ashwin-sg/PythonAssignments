#### 3. We want to buy paid advertisements on the website to promote our products. What is the best time to maximize the likelihood of customerâ€™s responding to the advertisement and purchasing the product?

```python
df2['Order_Date_New'] = pd.to_datetime(df2['Order Date'])
```

```python
df2['Hour'] = df2['Order_Date_New'].dt.hour

df2.head()
```

<h4>Answer: From the above fig. we can conclude that the peak time for maximum orders is around afternoon i.e. 10 to 12 Pm and in evening from 5 to 8 Pm. Hence, that should the time for paid advertisements </h4>

#### 4. What products are most often bought together by the customer?

```python
df2['Sold_Together'] = df2.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df2.head()
```

```python
df2 = df2[['Order ID','Sold_Together']].drop_duplicates()
```

```python
from itertools import combinations
from collections import Counter

sold_list = Counter()

for prod in df2['Sold_Together']:
    prod_list = prod.split(',')
    sold_list.update(Counter(combinations(prod_list,2)))
    
```

```python
sold_list.most_common(5)
```

<h4>Answer: The most often bought togehter products are iPhone and light charging cable</h4>
